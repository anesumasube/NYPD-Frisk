```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
clean_df = pd.read_csv('cleaned_df.csv', index_col=0)
clean_df.info()
```

```python
arrest_df = clean_df.loc[clean_df['SUSPECT_ARRESTED_FLAG'] == 'Y']
```

```python
arrest_df['SUSPECT_ARRESTED_FLAG'].value_counts()
```

```python
no_arrest_df= clean_df.loc[clean_df['SUSPECT_ARRESTED_FLAG'] == 'N']
```

```python

pal = sns.color_palette('muted')
sns.set_palette(pal)
sns.set_style('whitegrid')
sns.palplot(pal)
axis_font = {'fontname':'Arial', 'size':'14'}
title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} 
plt.figure(figsize=(10,7))
ax =sns.countplot(x='DAY2', data=arrest_df, color=pal[6])
plt.ylabel('Number of Arrests', **axis_font)
plt.xlabel('Day of the week', **axis_font)
plt.title("Arrest comparison by day", **title_font)
```

```python
unclean_df['Stop Frisk Time']= unclean_df
```

```python
start, stop = 0, 2
  
# converting to string data type 
unclean_df["Stop Frisk Time"]= unclean_df["Stop Frisk Time"].astype(str) 
  
# slicing till 2nd last element 
unclean_df["Hour"]= unclean_df["Stop Frisk Time"].str.slice(start, stop)
```

```python
unclean_df['Hour'] = unclean_df['Hour'].fillna(0)
```
